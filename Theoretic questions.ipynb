1) Почему для матрицы $X$ размера $m\times n$, при $m<n$ выполнено, что $X^TX$ - вырождена?

Дело в том, что ранг матирцы $X$ не больше чем $m$. 
То есть максимальное число оинейно независимых строчек, оно же максимальное число линейно независимых столбцов, оно же размерность образа оператора с матриицей $X$ не больше чем $m$.
Значит тем более ранг $X^TX$ не больше чем $m$ (размерность образа $X$ не больше чем разменость образа $YX$ для любых операторов $X, Y$).
Но у матрицы $X^TX$ $n$ строчек и столбцов, то есть они линейно зависимы, а значит матрица вырождена.

2) Если добавить регуляризацию, то матрица $Y = X^TX + \lambda I$ (тут $\lambda > 0$) - обратимая. Почему?

При $lambda > 0$ матрица $Y$ положительно определена, то есть для всех $v\neq0$: $(v^T, Yv) > 0$.
Это потому, что $(v^T, Yv)=\lambda(v^T, v) + (Xv; Xv) = \lambda||v||^2 + ||Xv||^2>0$.
Из положительноопределенности следует невырожденность, так как $Yx\neq 0$ при $v\neq0$.
Или просто у симметрической матрицы все собственные числа вещественны и неотрицательны, а теперь увеличены на $\lambda$.
Вообще регуляризация идет не по всем координатам, $\lambda$ не добавляется к коорденате соответствующей константе, но факт остается, ведь:
$(v^T, (X^TX + \lambda I_0)v)=\lambda(v^T_0, v_0) + (Xv; Xv) = \lambda||v_0||^2 + ||Xv||^2 > 0$, где $I_0$ это еденичная матрица, но с одним нулем, $v_0$ - $v$, но с нулем на первом месте. И либо $v_0\neq 0$, а значит $\lambda||v_0||^2>0$ либо у $v$ только первая координата ненулевая, но тогда $||Xv||^2 = X[1, 1]||v||^2 = n^2||v||^2 > 0$.
